####Desafio 10 - Tentar resolver o problema de plots de imagem antes de assistir o vídeo com a resolução Desafio 10 - Na tupla de retorno do forecast, estudar na documentação o que são cada um dos elementos de retorno

https://www.statsmodels.org/devel/generated/statsmodels.tsa.arima_model.ARIMAResults.forecast.html#statsmodels.tsa.arima_model.ARIMAResults.forecast

```python
from statsmodels.tsa.arima_model import ARIMA

dados_treino_desafio = dados_st.query("Ano != 2019")
dados_teste_desafio = dados_st.query("Ano == 2019").copy()

modelo_arima = ARIMA(dados_treino_desafio["Casos"].values.astype("float32"), order=[12,1,2])
modelo_arima_treinado = modelo_arima.fit()
```

```python
forecast, stderr, conf_int = modelo_arima_treinado.forecast(12)
```

```python
forecast
```

```python
forecast + stderr
```

```python
forecast - stderr
```

```python
conf_int
```

```python
dados_teste_desafio["forecast"] = forecast
dados_teste_desafio["forecast + stderr"] = forecast + stderr
dados_teste_desafio["forecast - stderr"] = forecast - stderr
dados_teste_desafio["conf_int0"] = [valor[0] for valor in conf_int]
dados_teste_desafio["conf_int1"] = [valor[1] for valor in conf_int]

plt.figure(figsize=(10,6))
ax = sns.lineplot(x="Data", y="forecast", data=dados_teste_desafio, label='previsão')
ax.fill_between(dados_teste_desafio['Data'], dados_teste_desafio["forecast - stderr"], dados_teste_desafio["forecast + stderr"], color='r', alpha=.1, label='Erro padrão da previsão')
ax.fill_between(dados_teste_desafio['Data'], dados_teste_desafio["conf_int0"], dados_teste_desafio["conf_int1"], color='g', alpha=.1, label='Intervalo de confiança da previsão')
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
ax.set_title('Entendendo o retorno do método fit do modelo ARIMA')
plt.show()
```

```python
from fbprophet import Prophet

modelo_prof = Prophet(seasonality_mode="multiplicative")
mapa = {"Data": "ds", "Passageiros": "y"}
dados_treino_passageiros_prof = dados_treino_passageiros[["Data", "Passageiros"]].rename(mapa,
                                                                                            axis=1)
modelo_prof_treinado = modelo_prof.fit(dados_treino_passageiros_prof)
```

```python
futuro = modelo_prof_treinado.make_future_dataframe(periods=12, freq="M")
futuro
```

```python
resultado_prof = modelo_prof_treinado.predict(futuro)
resultado_prof
```

```python
modelo_prof_treinado.plot(resultado_prof);
```
