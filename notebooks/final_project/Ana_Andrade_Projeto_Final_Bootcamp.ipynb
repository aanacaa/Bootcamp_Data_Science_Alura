COVID-19 é uma doença infeciosa causada pelo coronavírus da síndrome respiratória aguda grave 2 (SARS-CoV-2). Os sintomas mais comuns são febre, tosse seca e cansaço. Entre outros sintomas menos comuns estão dores musculares, dor de garganta, dor de cabeça, congestão nasal, conjuntivite, perda do olfato e do paladar e erupções cutâneas.

Cerca de 80% das infeções pelo SARS-CoV-2 confirmadas têm sintomas ligeiros de COVID-19 ou são assintomáticos, e a maioria recupera sem sequelas. Os casos mais graves podem evoluir para pneumonia grave com insuficiência respiratória grave, septicémia, falência de vários órgãos e morte.

Entre os sinais de agravamento da doença estão a falta de ar, dor ou pressão no peito, dedos de tom azul ou perturbações na fala e no movimento. O agravamento pode ser súbito, ocorre geralmente durante a segunda semana e requer atenção médica urgente.

````python
#Retorna as 5 primeiras linhas do dataset
dados.head()
````

````python
#Criei uma coluna chamada target, a qual vai receber o valor de ICU, mas posteriormente vou colocar se o paciente foi em qualquer uma das janelas para ICU.
dados['TARGET'] =  dados['ICU']
````

````python
dados.head(1000)
````

````python
#Set TARGET = 1 na janela 0-2 na caso o paciente tenha indo para UTI em alguma das outras linhas e retorna todas as linhas
def fn_prepare_window(rows):
    '''
    Set TARGET = 1 na janela 0-2 na caso o paciente tenha indo para UTI em alguma das outras linhas e retorna todas as linhas
    '''
    if(np.any(rows["TARGET"])):
        rows.loc[rows["WINDOW"]=="0-2", "TARGET"] = 1
    return rows
````

````python
#Aplica a função acima
dados = dados.groupby("PATIENT_VISIT_IDENTIFIER").apply(fn_prepare_window)
````

````python
dados.head(15)
````

Irei remover as linhas em que já é sinalizado a entrada do paciente na UTI,  sendo assim não corro risco de receber na janela 0-2 dados influenciados por resultados de exames após a entrada na UTI.

````python
#Removendo linhas
dados = dados.drop(dados.query("ICU==1").index)
````

````python
dados.head(20)
````

#### **Aplicação dos métodos bfill e ffill**

Conforme orientação do Sírio Libanês irei trabalhar com a janela de 0-2 para poder prever se o paciente irá ou não para UTI, entretanto temos vários dados null. Por isso, para preencher os dados será necessário adotar a estratégia do bfill e ffill.

````python
#Para poder ver todas as linhas que retorna abaixo:
pd.set_option('max_rows', 99999)
````

````python
#Procura dados null na janela de 0-2 e soma
dados[dados['WINDOW'] =='0-2'].isna().sum()
````

````python
def fn_preenche_tabela(dados):
    '''
    Agrupa pelo PATIENT_VISIT_IDENTIFIER e preenche os valores usando o bfill e ffill
    '''
    features_continuas_colunas = dados.iloc[:, 13:-2].columns
    features_continuas = dados.groupby("PATIENT_VISIT_IDENTIFIER", as_index=False)[features_continuas_colunas].fillna(method='bfill').fillna(method='ffill')
    features_categoricas = dados.iloc[:, :13]
    saida = dados.iloc[:, -2:]
    dados_finais = pd.concat([features_categoricas, features_continuas, saida], ignore_index=True,axis=1)
    dados_finais.columns = dados.columns
    return dados_finais
````

````python
#Prenchimendo de tabela
dados = fn_preenche_tabela(dados)
dados.head(15)
````

### **Função para validação cruzada**

````python
def fn_roda_modelo_cv(modelo, dados, n_splits, n_repeats):

    np.random.seed(1231234)
    dados = dados.sample(frac=1).reset_index(drop=True)
    x_columns = dados.columns
    y = dados["TARGET"]
    x = dados[x_columns].drop(["TARGET"], axis=1)
    
    
    cv = RepeatedStratifiedKFold(n_splits = n_splits, n_repeats=n_repeats)
    resultados=cross_validate(modelo, x, y, cv=cv, scoring='roc_auc', return_train_score=True)
    
    auc_medio = np.mean(resultados['test_score'])
    auc_medio_treino = np.mean(resultados['train_score'])

    auc_std = np.std(resultados['test_score'])
    
    print(f'AUC  {auc_medio} - {auc_medio_treino}')
    return auc_medio, auc_medio_treino
````

São os casos em que na realidade o paciente foi internado na UTI (TARGET = 1), mas que o modelo classificou incorretamente (TARGET Predicted = 0).

Portanto, conclui-se que essa área requer muito estudo de estatística e machine learning, além de um conhecimento prévio na área da saúde para desenvolver melhores classificadores.
